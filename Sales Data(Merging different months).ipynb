### Sales Analysis

<h4> This is a sales data for 12 months. Each month is in its excel workbook. Therefore, we will merging them together for our analysis

```python
# Library import
import pandas as pd
import os
 
```

<h4> Merging 12 months of sales data into a single file 

```python
# List all files in the directory
file_list = [file for file in os.listdir(r'C:\Users\NONSO\Desktop\Python for Data Analysis\DataSets\Sales_Data')]

# Print all filenames in the folder
for file in file_list:
    print(file)
```

```python
# Define the folder path
folder_path = r'C:\Users\NONSO\Desktop\Python for Data Analysis\DataSets\Sales_Data'

# Get all CSV file names in the folder
csv_files = [f for f in os.listdir(folder_path) if f.endswith('.csv')]

# Create an empty list to store DataFrames
df_list = []

# Loop through each CSV file and read it into a DataFrame
for file in csv_files:
    file_path = os.path.join(folder_path, file)  # Full file path
    df = pd.read_csv(file_path)  # Read CSV
    df_list.append(df)  # Append DataFrame to list

# Concatenate all DataFrames into one
df_combined = pd.concat(df_list, ignore_index=True)

# Display first few rows
print(df_combined.head())

# Save the merged DataFrame to a new CSV file (optional)
df_combined.to_csv(r'C:\Users\NONSO\Desktop\Python for Data Analysis\DataSets\Merged_Sales_Data.csv', index=False)
```

### What city has the highest number of sales?

```python
df_combined.head(5)          #returns the first 5 rows of the DataFrame
```

```python
# lets extract only the city from 'Purchase address'

def extract_city_and_state(address):
    # Split the address by commas
    address_parts = address.split(',')
    
    # Extract the city (second part) and state initials (third part)
    city = address_parts[1].strip()
    state_initials = address_parts[2].strip().split()[0]  # State is typically the first word in the third part
    
    # Combine city and state initials
    return f"{city} ({state_initials})"

# Apply the function to the 'Purchase Address' column to create a new 'City' column
df_combined['City'] = df_combined['Purchase Address'].apply(extract_city_and_state)


#df_combined['City'] = df_combined['Purchase Address'].apply(lambda x: x.split(',')[1])
#df_combined.head(2)
```

### What products are most often sold together?

```python
df_combined.head(3)         #returns the first 3 rows of the DataFrame    
```

```python
from itertools import combinations
from collections import Counter

# Load your dataset (assuming df_combined has 'Order ID' and 'Product' columns)
df_filtered = df_combined[df_combined['Order ID'].duplicated(keep=False)]  # Keep only orders with multiple items

# Group products by Order ID
df_grouped = df_filtered.groupby('Order ID')['Product'].apply(lambda x: ','.join(x)).reset_index()

# Display grouped products
print(df_grouped.head())
```

```python
# Create a Counter dictionary to count product pair occurrences
count_pairs = Counter()

# Iterate through each order and count product pairs
for products in df_grouped['Product']:
    product_list = products.split(',')  # Split products into a list
    product_combinations = list(combinations(product_list, 2))  # Create all possible 2-item combinations
    count_pairs.update(product_combinations)  # Update the Counter with these pairs

# Display the most common product pairs in a clear format
print("Most Frequently Bought Together Products:\n")
for pair, count in count_pairs.most_common(10):
    print(f"{pair[0]} & {pair[1]} --> {count} times")

# Convert top 10 product pairs to a DataFrame for better display
most_common_pairs = pd.DataFrame(count_pairs.most_common(10), columns=['Product Pair', 'Count'])

# Split tuple values into separate columns
most_common_pairs[['Product 1', 'Product 2']] = most_common_pairs['Product Pair'].apply(pd.Series)

# Drop the original tuple column
most_common_pairs.drop(columns=['Product Pair'], inplace=True)

# Display the DataFrame
#print("\nTop 10 Most Frequently Bought Together Products:")
#print(most_common_pairs)
```
